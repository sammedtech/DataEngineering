### Web Scrapping in Python

Scrapping data from online websites using their html tags

```python
# pip install bs4
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.flipkart.com/samsung-32-inch-4k-ultra-hd-led-backlit-va-panel-1-billion-colors-pbp-slim-bezels-flat-monitor-lu32j590uqwxxl/product-reviews/itm5b70dea04eb34?pid=MONFYGGAUFZAZKSZ&lid=LSTMONFYGGAUFZAZKSZWKSERF&marketplace=FLIPKART'
page = requests.get(url)
```

```python
page.text
```

```python
soup = bs(page.text, 'html.parser')
```

```python
soup.find_all('p')
```

```python
user_names = soup.find_all(class_ = "_2sc7ZR _2V5EHH")
```

```python
for name in user_names:
    names = str(name)
    names1 = names.replace('<p class="_2sc7ZR _2V5EHH">', '')
    names2 = names1.replace('</p>', '')
    print(names2)
```

```python
for name in user_names:
    print(name.get_text().strip())
```
