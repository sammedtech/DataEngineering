### Pandas Module
Pandas is mostly used by data engineers, data analysts and data scientists.

One of the most important library to explore as you'll be working 50% of time on the pandas library.

Used for data analysis purpose.

Used to work with large files.

Pandas Dataframes are 2-dimensional while pandas series is 1-dimensional

```python
import pandas as pd
```

```python
index = [1,2,3,4,5, 6]
Name = ["Jeevan", "Yogesh", "Sandeep", "Madhulika", "Prajakta", "Akshay"]
Age = [23, 24, 25, 26, 27, 28]
```

```python
# zip function : it'll map 2 different iterables based on their index
# ex. lst1 = [1,2,3], lst2 = ['a', 'b', 'c']
# list(zip(lst1, lst2)) --> [(1, 'a'), (2, 'b'), (3, 'c')]
emp_info = list(zip(index, Name, Age))
```

```python
emp_info
```

```python
# pd.DataFrame : To create new dataframe in pandas
emp_df = pd.DataFrame(emp_info, columns = ['S.No.', 'Name', 'Age'])
emp_df = emp_df.set_index(['S.No.'])
```

```python
type(emp_df)
```

```python
emp_df
```

```python
emp_df.loc[5, ['Name','Age']] 
```

```python
import numpy as np
```

```python
# Converting dataframe into numpy array
emp_arr = emp_df.to_numpy()
```

```python
type(emp_arr)
```

```python
emp_arr[0][1]
```

### Series Data Structure

Series data structure is basically a one-dimensional data structure

You can have one column as a series of data of any data type

```python
type(emp_df)
```

```python
type(emp_df.loc[:, 'Name'])
```

```python
emp_df.iloc[4, :]
```

#### How to access different elements of your DataFrame
df.loc --> column name with row index

df.iloc --> row index position

```python
emp_df.iloc[3]
```

```python
emp_df.loc[emp_df['Age'] > 25]
```
